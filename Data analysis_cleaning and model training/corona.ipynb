# Corona patients severity measure to be hospitalised


```python
import pandas as pd
```

```python
corona=pd.read_csv("corona.csv")
```

```python
corona.head()
```

```python
x=corona.drop('category',axis=1)
```

```python
x
```

```python
y=corona.category
```

```python
y
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
model_param={'svm':{'model':svm.SVC(gamma='auto'),'param':{'C':[1,5,10],'kernel':['rbf','linear', 'poly','sigmoid']}},
'RandomForestClassifier':{'model':RandomForestClassifier(max_depth=2, random_state=0),'param':{'n_estimators':[1,5,10,20,50,100]}},
'naive_bayes\(GuassianNB\)':{'model':GaussianNB(),'param':{ 'priors':[None]}},
'naive_bayes\(MultinomialNB\)':{'model':MultinomialNB(),'param':{'alpha':[1.0]}},
'DecisionTreeClassifier':{'model':DecisionTreeClassifier(),'param':{'criterion':['gini', 'entropy']}}} 
```

```python
from sklearn.model_selection import GridSearchCV
scores=[]
best_estimators = {}
for model_name,model_prop in model_param.items():
    model=GridSearchCV(model_prop['model'],model_prop['param'],cv=10)
    model.fit(x,y)
    scores.append({'model':model_name,'model_params':model.best_params_,'model_score':model.best_score_})
    best_estimators[model_name] = model.best_estimator_

best_estimators
```

```python
conclusion=pd.DataFrame(scores)
```

```python
conclusion
```

```python
best_model=best_estimators['svm']
```

```python
import joblib 
joblib.dump(best_model, 'saved_model.pkl') 
```
